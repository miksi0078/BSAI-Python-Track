Method 3 (enumerate)

```python
nums=[99,88,77]
for i, value in enumerate(nums):
    print(i, value)
```

```python
nums=[99,88,77,66,55]
for i, value in enumerate(nums,start=1):
    print(i, value)
```

LIST COMPREHENSION

```python
#traditional way
squares = []
for i in range(10):
    squares.append(i*i)
print(squares)
```

```python
#pythonic
squares = [i*i for i in range(10)]
print(squares)
```

```python
#with condition
evens = [i for i in range(20) if i%2==0]
print(evens)
```

DICTIONARIES

Structure:

key → value

```python
#Creating Dictionary
student = {
    "name": "Ali",
    "age": 21,
    "marks": 90
}
print(student)
```

Access Value

```python
print(student["name"])
print(student["marks"])
```

```python
#updatting
student["city"] = "swat"
print(student)
```

```python
for key in student:
    print(key, student[key])
```

```python
for key,value in student.items():
  print(key,value)
```

Nested Data Structures

Python allows structures inside structures.

```python
students = {
    "101": {"name":"Ali","marks":90},
    "102": {"name":"Sara","marks":95}
}
print(students["101"]["marks"])
```

```python
for key,value in students.items():
  print(key,value)
```

```python
students = {}

roll = input("Enter roll: ")
name = input("Enter name: ")
marks = int(input("Enter marks: "))

students[roll] = {
    "name": name,
    "marks": marks
}
print(students)
```
