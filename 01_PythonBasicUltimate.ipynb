```python
number = int(input("Enter number: "))

num = number
reverse = 0
digit_sum = 0

while num > 0:
    rem = num % 10
    reverse = reverse * 10 + rem
    digit_sum += rem
    num = num // 10

print("Reverse:", reverse)

if number == reverse:
    print("Palindrome: Yes")
else:
    print("Palindrome: No")

print("Sum of digits:", digit_sum)
```
