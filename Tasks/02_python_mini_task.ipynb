```python
#enhanced
students = {}
n = 2

for i in range(n):
    while True:
        try:
            roll = int(input(f"Enter roll number #{i+1}: "))
            break
        except ValueError:
            print("Invalid input. Roll number must be integer.")

    name = input(f"Enter name for roll {roll}: ")

    while True:
        try:
            marks = int(input(f"Enter marks for {name}: "))
            break
        except ValueError:
            print("Invalid input. Marks must be integer.")

    students[roll] = {"name": name, "marks": marks}

print("\nAll Students:")
for roll, info in students.items():
    print(f"Roll: {roll}, Name: {info['name']}, Marks: {info['marks']}")

# Find topper
topper_roll = max(students, key=lambda r: students[r]['marks'])
topper_info = students[topper_roll]

print(f"\nTopper: Roll {topper_roll}, Name {topper_info['name']}, Marks {topper_info['marks']}")
```

```python
def entry(n):
    students = {}

    for i in range(n):
        while True:
            try:
                roll = int(input(f"Enter roll number #{i+1}: "))
                break
            except ValueError:
                print("Invalid input. Roll number must be integer.")

        name = input(f"Enter name for roll {roll}: ")

        while True:
            try:
                marks = int(input(f"Enter marks for {name}: "))
                break
            except ValueError:
                print("Invalid input. Marks must be integer.")

        students[roll] = {"name": name, "marks": marks}

    return students


def findtopper(students):
    if not students:
        print("No students available.")
        return

    max_marks = -1
    topper_roll = None

    for roll, info in students.items():
        if info["marks"] > max_marks:
            max_marks = info["marks"]
            topper_roll = roll

    topper_info = students[topper_roll]

    print("\nTopper Details")
    print("Roll:", topper_roll)
    print("Name:", topper_info["name"])
    print("Marks:", topper_info["marks"])


students = {}

while True:

    print("\n1: Add Students")
    print("2: Find Topper")
    print("3: Exit")

    while True:
        try:
            choice = int(input("Enter choice please: "))
            break
        except ValueError:
            print("Invalid choice. Try again.")

    if choice == 1:
        try:
            n = int(input("How many students to add: "))
            students = entry(n)
        except ValueError:
            print("Invalid number.")

    elif choice == 2:
        findtopper(students)

    elif choice == 3:
        print("Program exiting...")
        break

    else:
        print("Invalid menu choice.")
```
