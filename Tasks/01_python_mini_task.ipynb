```python
def allinone():
    # Ask user how many numbers to input
    try:
        n = int(input("How many integers do you want to enter? "))
        if n <= 0:
            print("No numbers to process.")
            return {}
    except ValueError:
        print("Invalid input. Please enter an integer.")
        return {}

    # Collect numbers
    numbers = []
    for i in range(n):
        while True:
            try:
                temp = int(input(f"Enter integer #{i+1}: "))
                numbers.append(temp)
                break
            except ValueError:
                print("Invalid input. Please enter an integer.")

    print("\nOriginal List:", numbers)

    # Separate even and odd numbers
    even_numbers = [i for i in numbers if i % 2 == 0]
    odd_numbers = [i for i in numbers if i % 2 != 0]

    even_numbers.sort()
    odd_numbers.sort()

    # Initialize sum, max, min
    total = 0
    maximum = numbers[0]
    minimum = numbers[0]

    for num in numbers:
        total += num
        if num > maximum:
            maximum = num
        if num < minimum:
            minimum = num

    # Prepare structured result
    result = {
        "original_list": numbers,
        "even_numbers": even_numbers,
        "odd_numbers": odd_numbers,
        "sum": total,
        "maximum": maximum,
        "minimum": minimum
    }

    return result

# Call function
result = allinone()
print("\nProcessed Result:")
for key, value in result.items():
    print(f"{key}: {value}")
```
