Create list of numbers.

Remove duplicates without using set.

```python
def removeduplicate(numbers):
    temp = []

    for num in numbers:
        if num not in temp:
            temp.append(num)

    return temp
number=[]
while True:
  try:
    size=int(input("Enter size of list:\t"))
    break
  except ValueError:
    print("inalid input try angain!")
for i in range(size):
  while True:
    try:
       temp=int(input("Enter any integer eg(1,2):\t"))
       break
    except ValueError:
      print("Enter valid integer!")
  number.append(temp)
  temp=0
print(number)
print(removeduplicate(number))



```

TASK 5 (Algorithmic)

Given list:

[4,7,2,9,1]

Find second largest element.

Without using sort().

```python
num=[4,7,2,9,1]
largest=num[0]
for i in range(len(num)):
  if num[i]>largest:
    largest=num[i]
print(largest)
second=num[0]
for j in range(len(num)):
  if num[j]>second and num[j]<largest:
    second=num[j]
print(second)
```

```python
num=[4,7,2,9,1]
num.sort()
print(num[-1])
print(num[-2])
```

```python
num=[4,7,2,9,1]
num.sort(reverse=True)
print(num[0])
print(num[1])
```

```python
num = [4,7,2,9,1]

largest = float('-inf') #neg infinity
second = float('-inf')

for n in num:

    if n > largest:
        second = largest
        largest = n

    elif n > second and n != largest:
        second = n

print("Largest:", largest)
print("Second Largest:", second)
```
