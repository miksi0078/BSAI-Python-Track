```python
#Basic Function
def greet():
    print("Hello Miksi")

greet()
```

```python
#function with parameter
#name is parameter hwere
def greet(name):
    print("Hello", name)

greet("Ali")
```

```python
#returning value
def add(a, b):
    return a + b

result = add(5, 3)
print(result)
```

```python
#multipile returns
def operations(a, b):
    return a+b, a-b, a*b

x, y, z = operations(10, 5)
print(f"{x} {y} {z}")
```

```python
#exp is defult if not provided auto 2 used but base is required
def power(base, exp=2):
    return base ** exp

print(power(5))
print(power(5, 3))
```

```python
#key arguments
def student(name, marks):
    print(name, marks)

student(marks=90, name="Miksi")
```

```python
#varable lenghth arguments
def total(*numbers):
    print(numbers)
    return sum(numbers)

print(total(10, 20, 30))
```

```python
#Lambda Functions
square = lambda x: x*x
print(square(5))
#lambda parameters : expression
"""
Important:

It can take multiple parameters

It can only contain one expression

That expression is automatically returned

No return keyword
You need a small temporary function

You don’t want to formally define it with def

You pass functions as arguments


"""
```
