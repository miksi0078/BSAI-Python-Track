```python
existing_models = ["Mood_Tracker_AI", "ParkEase_Vision"]
existing_accuracies = [88.5, 92.0]

print("--- PHASE 1: The String Gate ---")
i = 0
phase1_success = False
valid_name = "" # Temporary storage

while i < 3:
    temp_name = input("Enter new AI model name (min 4 chars): ")

    if len(temp_name) >= 4:
        valid_name = temp_name # Store, but do not append yet
        phase1_success = True
        break
    else:
        print("Invalid length. Must be at least 4 characters.")
        i += 1

if not phase1_success:
    print("Max attempts reached. System locked.")
else:
    print("\n--- PHASE 2: The Float Gate ---")
    j = 0
    phase2_success = False
    valid_acc = 0.0 # Temporary storage

    while j < 3:
        try:
            temp_acc = float(input(f"Enter accuracy for {valid_name} (in float): "))
            valid_acc = temp_acc # Store, but do not append yet
            phase2_success = True
            break
        except ValueError:
            print("Invalid input. Accuracy must be a numerical value.")
            j += 1

    if not phase2_success:
        print("Max accuracy attempts reached. Pipeline terminated.")
    else:
        print("\n--- PHASE 3: Data Integration ---")
        # Only append when BOTH inputs are verified. This protects parallel structure.
        existing_models.append(valid_name)
        existing_accuracies.append(valid_acc)

        # Convert to permanent Dictionary storage
        pipeline_dict = dict(zip(existing_models, existing_accuracies))
        print(f"Current Pipeline Matrix:\n{pipeline_dict}")

        print("\n--- PHASE 4: Final Evaluation Report ---")
        # Iterate through the parallel lists dynamically
        for name, accuracy in zip(existing_models, existing_accuracies):
            if accuracy >= 90.0:
                print(f"[PASS] {name} is cleared for production. ({accuracy}%)")
            elif accuracy >= 75.0:
                print(f"[TUNE] {name} requires optimization. ({accuracy}%)")
            else:
                print(f"[FAIL] {name} failed baseline. ({accuracy}%)")
```
