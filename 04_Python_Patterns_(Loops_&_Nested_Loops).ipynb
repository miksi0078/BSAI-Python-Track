```python
for i in range(1,10):
  if i%2==0:
    print("even :",i)
  else:
    print("odd; ", i)
```

```python
for i in range(1,6):
    for j in range(1,i+1):
        print(j, end=" ")
    print() #new line
```

```python
n = 5
for i in range(1, n+1):
    print(" " * (n - i) + "*" * (2*i - 1))
```

```python
n = 5
for i in range(1, n+1):
    for j in range(1, 2*n):
        if j == n-i+1 or j == n+i-1 or i == n:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```

```python
n = 5  # number of rows

for i in range(1, n+1):
    # print leading spaces
    print(" " * (n - i), end="")

    # print even numbers
    for j in range(1, i+1):
        print(2*j, end=" ")

    print()  # new line
```

```python
n = 5  # middle row

# Upper pyramid
for i in range(1, n+1):
    print(" " * (n - i) + "*" * (2*i - 1))

# Lower inverted pyramid
for i in range(n-1, 0, -1):
    print(" " * (n - i) + "*" * (2*i - 1))
```

```python
n = 5

for i in range(n, 0, -1):  # rows decreasing
    for j in range(1, i+1):  # numbers in each row
        print(j, end=" ")
    print()  # new line
```
