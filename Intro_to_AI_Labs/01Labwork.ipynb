```python
# 1. Inputs and Calculations
sell_price = float(input("Enter sell price: "))
profit = sell_price * 0.25
cost_price = sell_price - profit
name = input("Enter product name: ")

profit_percentage = (profit / cost_price) * 100

# 2. Correct Dictionary Syntax
# A dictionary uses curly braces {} and key: value pairs[cite: 86].
product = {
    "Name": name,
    "Selling Price": sell_price,
    "Cost Price": cost_price,
    "Profit": profit,
    "Profit Percentage": profit_percentage
}

print("\n--- Product Details ---")

# 3. Iteration and Type Checking
# We use .items() to get both the label and the data.
for key, value in product.items():
    # Logic: If the value is text (str), print it normally to avoid ValueError.
    if type(value) == str:
        print(f"{key}: {value}")
    # Otherwise, it is a number, so we can safely apply :.2f formatting.
    else:
        print(f"{key}: {value:.2f}")
```

```python
9# Take input
num_1 = int(input("Enter first number: "))
num_2 = int(input("Enter second number: "))

# Basic operations
addition = num_1 + num_2
subtraction = num_1 - num_2
multiplication = num_1 * num_2
power = num_1 ** num_2

# Division-related operations using EAFP principle
try:
    division = num_1 / num_2
    floor_division = num_1 // num_2
    modulus = num_1 % num_2
except ZeroDivisionError:
    division = floor_division = modulus = None

# Store results in dictionary
results = {
    "Addition": addition,
    "Subtraction": subtraction,
    "Multiplication": multiplication,
    "Division": division,
    "Floor Division": floor_division,
    "Modulus": modulus,
    "Power": power
}

# Display results
for operation, value in results.items():
    if value is None:
        print(f"{operation}: Not Possible (Division by Zero)")
    else:
        print(f"{operation}: {value}")
```
