```python
# Write a program to check whether a number is prime.
# Note: Take input from user.

def is_prime(n):
    if n <= 1:
        print("Enter a number greater than 1.")
        return False

    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

limit = 0
num_1 = None

while limit < 3:
    limit += 1
    try:
        num_1 = int(input("Enter number greater than 1 please:\t"))
        break
    except ValueError:
        print("Please enter some valid input... eg 1, 2")

if num_1 is not None:
    print("Number is prime:\t", is_prime(num_1))
else:
    print("Max limits reached terminating program goodbye!")
```

```python
#Take input from user and create a function to compute its factorial using a loop
## finding factorials
def finding_fctorial(n):
    if n < 0:
        return "number cant be negative less than zero(0)"
    sum = 1
    for i in range(2, n+1):
        sum = i * sum
    return sum

try:
    num_1 = int(input("enter number please..."))
    print("factorial is:", finding_fctorial(num_1))
except ValueError: # Fix 1: Corrected spelling
    print("invalid input>>>")
```

```python
# Write a program to count vowels in a string.
def counting_vowels(strr):
    count = 0
    for ch in strr:
        if ch in 'aeiouAEIOU':
            count += 1
    return count
sentence = input("Enter  a sentence please...: ")
print("Vowels are :\t", counting_vowels(sentence))
```

```python
#alternative
# Write a program to count vowels in a string.
def counting_vowels(strr):
    count = 0
    for ch in strr:
        if (ch == 'a' or ch == 'A' or ch == 'e' or ch == 'E' or
            ch == 'i' or ch == 'I' or ch == 'o' or ch == 'O' or
            ch == 'u' or ch == 'U'):
            count += 1
    return count
sentance = input("Enter sentence: ")
print("Vowels are :\t", counting_vowels(sentance))
```

```python
# Task 1: Create a tuple of 5 students
student_records = (
    (101, "khuram", (80, 75, 85)),
    (102, "sofi", (90, 92, 88)),
    (103, "miksi", (60, 65, 70)),
    (104, "artiza", (91, 92, 88)),
    (105, "baryal yousafzai", (76, 78, 80))
)


high_avg = 0.0
top_student = ""
above_75 = 0
print("printing studens averages...")
for student in student_records:
    name = student[1]
    marks = student[2]
    total_marks = marks[0] + marks[1] + marks[2]
    student_avg = total_marks / 3.0
    print(name, "average is:", student_avg)

    if student_avg > 75:
        above_75 = above_75 + 1
    if student_avg > high_avg:
        high_avg = student_avg
        top_student = name


print("Top student is\t:", top_student, "with score is:\t", high_avg)
print("Total above 75 student are\t:", above_75)


print("updating...")
temp_list = list(student_records)
temp_list[2] = (103, "miksi", (80, 85, 90))
updated_records = tuple(temp_list)

for data in updated_records:
    print(data)
```
