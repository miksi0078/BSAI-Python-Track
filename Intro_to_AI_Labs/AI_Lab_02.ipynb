```python
value = int(input("Enter a value: "))

original = value

results = {}

temp = original
temp += 2
results["+= 2"] = temp

temp = original
temp -= 2
results["-= 2"] = temp

temp = original
temp *= 2
results["*= 2"] = temp

temp = original
temp /= 2
results["/= 2"] = temp

for operation, result in results.items():
    print(f"Original value: {original} | After {operation} → {result}")
```

```python
def decimal_to_binary(decimal):
    """
    Convert a decimal number (<1024) to binary.

    Parameters:
        decimal (int): decimal number, must be <1024

    Returns:
        str: binary representation (without '0b' prefix)
    """
    if not (0 <= decimal < 1024):
        return "Error: Number must be between 0 and 1023"

    return bin(decimal)[2:]  # remove '0b' prefix of the function bin

# Example usage
try:
    num = int(input("Enter a decimal number less than 1024: "))
except ValueError:
    print("Invalid input! Please enter an integer.")
    exit()

binary_num = decimal_to_binary(num)
print(f"Decimal {num} → Binary {binary_num}")
```
