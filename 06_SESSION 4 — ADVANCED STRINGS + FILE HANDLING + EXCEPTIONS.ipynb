Mini-Project: Word Counter

```python
sentence = "AI is powerful and AI is future"
words = sentence.split()
freq = {}
for w in words:
    if w in freq:
        freq[w] += 1
    else:
        freq[w] = 1
print(freq)
```

Exception Handling in Python

```python
try:
    x = int(input("Enter a number: "))
except ValueError:
    print("Invalid input!")
```

```python
try:
    a = int(input("A: "))
    b = int(input("B: "))
    c = a / b
except ValueError:
    print("Please enter integers only")
except ZeroDivisionError:
    print("Division by zero is not allowed")
```

Finally Block

```python
try:
    file = open("data.txt","r")
    content = file.read()
    print(content)
except FileNotFoundError:
    print("File not found")
finally:
    print("Execution finished")
```

Task: Word Frequency Counter from File

```python
with open("story.txt", "w") as f:
    f.write("""The Adventures of Luna the Cat

Luna was a small, curious cat with bright green eyes. Every morning, she would climb to the top of the old wooden fence and watch the world wake up. Birds chirped, leaves rustled, and the sun painted golden streaks across the garden.

One day, Luna noticed a shiny object near the pond. She tiptoed carefully and discovered a tiny, sparkling key. "What could this open?" she wondered. Her heart raced with excitement.

For days, Luna searched the garden and the old house, trying the key on every lock. Just as she was about to give up, she found a small door hidden behind the ivy. With a gentle turn, the key clicked, and the door creaked open.

Inside, Luna discovered a secret room filled with cozy pillows, twinkling lights, and tiny treats. It became her favorite place to hide, dream, and plan her next adventure. From that day on, Luna knew that curiosity always leads to magical discoveries.
""")
try:
    with open("story.txt","r") as f:
        text = f.read()
        print(text)
except FileNotFoundError:
    print("story.txt not found")
    text = ""

if text:
    words = text.split()
    freq = {}
    for w in words:
        freq[w] = freq.get(w,0) + 1

    print("Word frequency:")
    for word,count in freq.items():
        print(f"{word} : {count}")
```

words = ['hi','hello','hi']
freq = {}

# First word: 'hi'
freq.get('hi',0) → 0
freq['hi'] = 0 + 1 → freq = {'hi':1}

# Second word: 'hello'
freq.get('hello',0) → 0
freq['hello'] = 0 + 1 → freq = {'hi':1, 'hello':1}

# Third word: 'hi'
freq.get('hi',0) → 1
freq['hi'] = 1 + 1 → freq = {'hi':2, 'hello':1}
