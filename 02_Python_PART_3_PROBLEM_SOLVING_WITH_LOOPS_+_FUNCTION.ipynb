```python
def is_armstrong(n):
    # Reject non-integers
    if not isinstance(n, int):
        return False

    # Reject negative numbers
    if n < 0:
        return False

    # Special case: 0 is an Armstrong number
    if n == 0:
        return True

    original = n
    digit_count = 0

    # Count digits
    temp = n
    while temp > 0:
        temp //= 10
        digit_count += 1

    power_sum = 0
    temp = n

    # Calculate sum of digits raised to digit_count
    while temp > 0:
        digit = temp % 10
        power_sum += digit ** digit_count
        temp //= 10

    return original == power_sum
print("153 is argonstrong:\t",is_armstrong(153))
```
