```python
# 1. Creation: Making a list of three integers
system_codes = [100, 200, 300]

# 2. Reading: Accessing the first item (Index 0)
first_code = system_codes[0]

# 3. Updating: Changing the first item from 100 to 150
system_codes[0] = 150

# 4. Adding: Pushing a new item to the end of the list
# (This is Python's version of C++ push_back)
system_codes.append(400)

# 5. Output: Printing the entire list to see the final result
print(system_codes)

# The output in the terminal will be: [150, 200, 300, 400]
```

```python
ml_accuracies = [90, 78, 66, 88]
ml_names = ["drivesef", "buyself", "clearningai", "Bpmonoteringai"]

# The zip function pairs the 0th index of both lists, then the 1st, and so on.
for name, accuracy in zip(ml_names, ml_accuracies):
    if accuracy >= 90:
        print(f"{name} has passed the test with accuracy: {accuracy}")
    else:
        print(f"{name} has failed the test with accuracy: {accuracy}")
```

```python
# 1. Your original parallel lists
ml_names = ["drivesef", "buyself", "clearningai", "Bpmonoteringai"]
ml_accuracies = [90, 78, 66, 88]

print("--- PART 1: The zip() Loop ---")
# Using zip() to loop through both lists at the exact same time
for name, accuracy in zip(ml_names, ml_accuracies):
    if accuracy >= 90:
        print(f"{name} has passed the test with accuracy: {accuracy}")
    else:
        print(f"{name} has failed the test with accuracy: {accuracy}")

print("\n--- PART 2: The dict(zip()) Conversion ---")
# Using zip() to instantly combine two lists into one Dictionary
model_performance = dict(zip(ml_names, ml_accuracies))

# Printing the whole dictionary to see the Key:Value structure
print(f"The combined dictionary looks like this:\n{model_performance}")

# Extracting one specific value using its Key (no loops needed)
print(f"\nDirect query for clearningai accuracy: {model_performance['clearningai']}")
```

```python
courses = ["Probability", "Data Structures", "Machine Learning", "Linear Algebra"]
scores = [78, 85, 92, 65]
future_scope_probability = [0.8, 0.5, 0.7, 0.7]

# Drop the dict(), just use zip() for parallel iteration
for course, score, prob in zip(courses, scores, future_scope_probability):

    # Use 'and' instead of '&&'
    if prob >= 0.7 and score >= 70:
        print(f"Course name: {course} has passed test with score {score} and future probability {prob}")
    else:
        print(f"Course name: {course} has failed test with score {score} and future probability {prob}")
```
