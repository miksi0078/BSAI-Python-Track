```python
try:
    # Python attempts this risky line
    user_age = int(input("Enter age: "))
    print(f"You are {user_age}")
except ValueError:
    # If the user typed text, the program catches the crash here
    print("Error: You must enter a number, not text.")
```

```python
# The loop starts and will run forever until it hits a 'break'
while True:
    try:
        # 1. Attempt the risky conversion
        user_age = int(input("Enter your age: "))

        # 2. If the conversion succeeds, the code reaches this line.
        # We break the infinite loop because we got valid data.
        break

    except ValueError:
        # 3. If the user types text, float() crashes and jumps here IMMEDIATELY.
        # It skips the 'break' command.
        # The error prints, and the while loop restarts from the top.
        print("Invalid input. You must type a number. Try again.\n")

# 4. The program only reaches here once the loop is broken.
print(f"Age safely captured: {user_age}")
```

```python
i = 0
user_age = None  # Pre-declare the variable with an empty value

while True:
    if i < 4:
        try:
            user_age = int(input("Enter your age: "))
            break  # Success: break the loop

        except ValueError:
            print("Invalid input. You must type a number. Try again.")
            i += 1  # BUG 1 FIXED: Increment the attempt counter

    else:
        print("Max attempts reached.")
        break

# BUG 2 FIXED: Only print the age if the variable was actually updated
if user_age is not None:
    print(f"Age safely captured: {user_age}")
else:
    print("Program terminated due to invalid input.")
```

```python
i = 0
user_age = None

# The loop automatically stops if i reaches 4
while i < 4:
    try:
        user_age = int(input("Enter your age: "))
        break  # Success
    except ValueError:
        print("Invalid input. Try again.")
        i += 1  # Increment on failure

if user_age is not None:
    print(f"Age safely captured: {user_age}")
else:
    print("Max attempts reached.")
```

```python
i = 0
algo_name = None

while i < 4:
    # 1. Capture the input (No try block needed because strings don't crash here)
    temp_name = input("Enter AI name please (min 5 chars): ")

    # 2. Validate the data using logic, not error handling
    if len(temp_name) >= 5:
        algo_name = temp_name  # Save the valid name
        break                  # Break the loop
    else:
        print("Invalid input. String must be at least 5 characters.")
        i += 1                 # Increment the counter on failure

if algo_name is not None:
    print(f"Algorithm name safely captured: {algo_name}")
else:
    print("Max attempts reached.")
```
